```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
Summary_df = pd.DataFrame()

# Total Schools, add to Summary DF 
TotalSchools = pd.Series(school_data_complete['School ID'].nunique())
Summary_df['Total Number of Schools'] = TotalSchools

# Total Students, add to Summary DF 
TotalStudents = school_data_complete['Student ID'].nunique()
Summary_df['Total Number of Students'] = TotalStudents

# Total Budget, add to Summary DF 
TotalBudget = school_data_complete['budget'].unique().sum()
Summary_df['Total Schools budget'] = TotalBudget

# Calculate Reading and Maths Averages, add to Summary DF 
AverageMathScore = school_data_complete['math_score'].mean()
Summary_df['Average Math Score'] = AverageMathScore
AverageReadingScore = school_data_complete['reading_score'].mean()
Summary_df['Average Reading Score'] = AverageReadingScore
NumStudentsPassingMaths = school_data_complete['math_score'][school_data_complete["math_score"] >= 70].count()

# Calculate percentage of students passing reading and maths, add to summary DF 
NumStudentsPassingReading = school_data_complete['reading_score'][school_data_complete["reading_score"] >= 70].count()
NumStudentsPassingMathsAndReading = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].count()
PctStudentsPassingMaths = NumStudentsPassingMaths / TotalStudents
PctStudentsPassingReading = NumStudentsPassingReading / TotalStudents
PctStudentsPassingMathsAndReading = NumStudentsPassingMathsAndReading / TotalStudents
Summary_df['% Students Passing Maths'] = PctStudentsPassingMaths
Summary_df['% Students Passing Reading'] = PctStudentsPassingReading
# below variable passed as a series not a value, extract a value 
Summary_df['% Students Passing Maths & Reading'] = PctStudentsPassingMathsAndReading['reading_score']

Summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# create empty dataframe 
School_Summary_df = pd.DataFrame()

# create GroupBy table, group by School
schools_groupby_df = school_data_complete.groupby('school_name')

# Get Total Students per School
School_Summary_df['Students per School'] = schools_groupby_df['Student ID'].nunique()

# Get School Budget, get mean as values are same 
School_Summary_df['Budget per School'] = schools_groupby_df['budget'].mean()

# Calculate Per Student Budget
School_Summary_df['Budget per Student'] = School_Summary_df['Budget per School'] / School_Summary_df['Students per School']

# Calculate Average Math Score 
School_Summary_df['Average Math Score'] = schools_groupby_df['math_score'].mean()

# Calculate Average Reading Score 
School_Summary_df['Average Reading Score'] = schools_groupby_df['reading_score'].mean()

# Calc Students Passing Maths 
NumStudentsPassingMaths = (school_data_complete['math_score'] >=70).groupby(school_data_complete.school_name).sum()
School_Summary_df['% Students Passing Maths'] = NumStudentsPassingMaths / School_Summary_df['Students per School']

# Calc Students Passing Reading
NumStudentsPassingReading = (school_data_complete['reading_score'] >=70).groupby(school_data_complete.school_name).sum()
School_Summary_df['% Students Passing Reading'] = NumStudentsPassingReading / School_Summary_df['Students per School']

# Calc Students Passing Maths and Reading
NumStudentsPassingMathsAndReading = ((school_data_complete['reading_score'] >=70) & (school_data_complete['math_score'] >=70)).groupby(school_data_complete.school_name).sum()
School_Summary_df['% Students Passing Maths and Reading'] = NumStudentsPassingMathsAndReading / School_Summary_df['Students per School']

School_Summary_df
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
# Clone Previous DataFrame 
Top_Performing_Schools_df = School_Summary_df.copy()

Top_Performing_Schools_df = Top_Performing_Schools_df.sort_values(by='% Students Passing Maths and Reading', ascending = False)
Top_Performing_Schools_df = Top_Performing_Schools_df.nlargest(5, '% Students Passing Maths and Reading')
Top_Performing_Schools_df
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
Bottom_Performing_Schools_df = School_Summary_df.copy()

Bottom_Performing_Schools_Sorted_df = Bottom_Performing_Schools_df.sort_values(by='% Students Passing Maths and Reading', ascending = True)
Bottom_Performing_Schools_Sorted_df = Bottom_Performing_Schools_df.nsmallest(5, '% Students Passing Maths and Reading')

Bottom_Performing_Schools_Sorted_df
```
